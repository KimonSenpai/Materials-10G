```python
def Abs(f):
    def ff(x):
        return abs(f(x))
    return ff

def Add(f, a):
    def ff(x):
        return f(x) + a
    return ff

def Mul(f, a):
    def ff(x):
        return f(x) * a
    return ff
```

```python
def Ind(x):
    return x
f = Abs(Add(Abs(Add(Mul(Abs(Add(Ind, 5)), 2), -2)), +3))
Y = [f(x) for x in X]
mp.plot(X, Y)
mp.show()
```

```python
def gen_g(X, i):
    mas = list(X)
    def g(x):
        numerator = 1
        denominator = 1
        for j in range(len(mas)):
            if j == i: continue
            numerator *= (x - mas[j])
            denominator *= (mas[i] - mas[j])
        return numerator/denominator
    
    return g

def gen_f(X, Y):
    X = list(X)
    Y = list(Y)
    G = []
    for i in range(len(X)):
        G += [gen_g(X, i)]
    def f(x):
        s = 0
        for y, g in zip(Y, G):
            s += y*g(x)
        return s
    return f
```

```python
X = [-2, 0, 4, 5]
Y = [0, 2, 0, 3]
f = gen_f(X, Y)

XX = [val/100 for val in range(-200, 500)]
YY = [f(x) for x in XX]
mp.plot(XX, YY)
mp.show()
```
