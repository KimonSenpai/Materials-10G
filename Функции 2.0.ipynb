```python
import matplotlib.pyplot as mp
X = [v/100 for v in range(-2000, 2000)]
Y = [v*v for v in X]
mp.plot(X, Y)
mp.show()
```

```python
def b(x):
    return x**2

c = b
print(c(2))
```

```python
def gen_f(a):
    def f(x):
        return x**a
    return f

g = gen_f(3)
Y2 = [g(x) for x in X]
mp.plot(X, Y)
mp.plot(X, Y2)
mp.show()
```

```python
def Abs(f):
    def ff(x):
        return abs(f(x))
    return ff

def Add(f, a):
    def ff(x):
        return f(x) + a
    return ff

def Mul(f, a):
    def ff(x):
        return f(x) * a
    return ff
```

```python
def Ind(x):
    return x
f = Abs(Add(Abs(Add(Mul(Abs(Add(Ind, 5)), 2), -2)), +3))
Y = [f(x) for x in X]
mp.plot(X, Y)
mp.show()
```
