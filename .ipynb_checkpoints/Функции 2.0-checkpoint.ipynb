```python
import matplotlib.pyplot as mp
X = [v/100 for v in range(-500, 500)]
Y = [v*v for v in X]
mp.plot(X, Y)
mp.show()
```

```python
def b(x):
    return x**2

c = b
print(c(2))
```

```python
def gen_f(a):
    def f(x):
        return x**a
    return f

g = gen_f(3)
Y2 = [g(x) for x in X]
mp.plot(X, Y)
mp.plot(X, Y2)
mp.show()
```
